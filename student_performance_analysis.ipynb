```python
print(df.columns)
```

```python
df[['mark_arabic', 'mark_english', 'mark_math', 'mark_mechanic', 'mark_physics', 'mark_social']] = \
    df[['mark_arabic', 'mark_english', 'mark_math', 'mark_mechanic', 'mark_physics', 'mark_social']].apply(pd.to_numeric, errors='coerce')

# Calculate the total marks by summing across the selected columns
df['total_marks'] = df[['mark_arabic', 'mark_english', 'mark_math', 'mark_mechanic', 'mark_physics', 'mark_social']].sum(axis=1)

# Get the top 10 students based on the total marks
top_students = df.nlargest(10, 'total_marks')

# Display the result
print(top_students[['names', 'total_marks']])
```

```python
# Bar Plot of Total Marks
plt.figure(figsize=(12, 8))
sns.barplot(x='names', y='total_marks', data=top_students)
plt.xticks(rotation=90)
plt.title('Top 10 Students by Total Marks')
plt.xlabel('Names')
plt.ylabel('Total Marks')
plt.show()
```

```python
df.replace("غ", pd.NA, inplace=True)

# Convert marks to numeric values, coerce errors (convert invalid entries to NaN)
df[['mark_arabic', 'mark_english', 'mark_math', 'mark_mechanic', 'mark_physics', 'mark_social']] = df[['mark_arabic', 'mark_english', 'mark_math', 'mark_mechanic', 'mark_physics', 'mark_social']].apply(pd.to_numeric, errors='coerce')
```

```python
avg_marks = df[['mark_arabic', 'mark_english', 'mark_math', 'mark_mechanic', 'mark_physics', 'mark_social']].mean()

# Plot
plt.figure(figsize=(10, 6))
avg_marks.plot(kind='bar', color='skyblue')
plt.title('Average Marks by Subject')
plt.ylabel('Average Marks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['mark_arabic', 'mark_english', 'mark_math', 'mark_mechanic', 'mark_physics', 'mark_social']])
plt.title('Distribution of Marks by Subject')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define a function to categorize grades into "Good", "Bad", or "Average"
def categorize_performance(grade):
    if isinstance(grade, (int, float)):
        if grade >= 15:
            return 'Good'
        elif grade < 10:
            return 'Bad'
        else:
            return 'Average'
    return 'Unknown'  # For missing or non-numeric grades like "غ"

# Apply this categorization function for each subject
df['performance_arabic'] = df['mark_arabic'].apply(categorize_performance)
df['performance_math'] = df['mark_math'].apply(categorize_performance)
df['performance_physics'] = df['mark_physics'].apply(categorize_performance)
df['performance_social'] = df['mark_social'].apply(categorize_performance)
df['performance_mechanic'] = df['mark_mechanic'].apply(categorize_performance)
df['performance_english'] = df['mark_english'].apply(categorize_performance)

# Check the changes in the data
df[['names', 'performance_arabic', 'performance_math', 'performance_physics', 'performance_social', 'performance_mechanic','performance_english']].head()
```

```python
df_melted = df.melt(id_vars=['names'], value_vars=['performance_arabic', 'performance_math', 'performance_physics', 'performance_social', 'performance_mechanic','performance_english'],
                    var_name='subject', value_name='performance')

# Create the count plot to show the performance distribution across subjects
plt.figure(figsize=(12, 8))
sns.countplot(x='subject', hue='performance', data=df_melted, palette='Set2')
plt.title('Performance Distribution Across Subjects')
plt.xlabel('Subject')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```
