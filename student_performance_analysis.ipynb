```python
# Function to categorize performance
def categorize_performance(grade, full_mark):
    """
    Categorize performance based on grade and full mark.
    - "Excellent": 85% or above of the full mark
    - "Good": between 70% and 85% of the full mark
    - "Weak": between 50% and 70% of the full mark
    - "Bad": below 50% of the full mark
    """
    try:
        # Ensure the grade is numeric by converting it to float
        grade = float(grade)
        if grade >= 0.85 * full_mark:
            return 'Excellent'
        elif 0.7 * full_mark <= grade < 0.85 * full_mark:
            return 'Good'
        elif 0.5 * full_mark <= grade < 0.7 * full_mark:
            return 'Weak'
        else:
            return 'Fail'
    except (ValueError, TypeError):  # Catch non-numeric grades
        return 'Unknown'  # For invalid grades or missing values

# Apply this categorization function for each subject with their respective full marks
df['performance_arabic'] = df['mark_arabic'].apply(lambda x: categorize_performance(x, 25))
df['performance_math'] = df['mark_math'].apply(lambda x: categorize_performance(x, 30))
df['performance_physics'] = df['mark_physics'].apply(lambda x: categorize_performance(x, 25))
df['performance_social'] = df['mark_social'].apply(lambda x: categorize_performance(x, 25))
df['performance_mechanic'] = df['mark_mechanic'].apply(lambda x: categorize_performance(x, 30))
df['performance_english'] = df['mark_english'].apply(lambda x: categorize_performance(x, 20))

# Melt the dataframe for easier plotting
df_melted = df.melt(id_vars=['names'], 
                    value_vars=['performance_arabic', 'performance_math', 'performance_physics', 'performance_social', 
                                'performance_mechanic', 'performance_english'],
                    var_name='subject', value_name='performance')
```
